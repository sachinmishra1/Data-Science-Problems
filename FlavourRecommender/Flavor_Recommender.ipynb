```python
from sklearn.preprocessing import LabelEncoder
Encd = LabelEncoder()
Encd.fit(['Male','Female'])
```

```python
df['Gender'] = Encd.transform(df['Gender'])
```

```python
df.head(10)
```

```python
x = df.drop(columns = ['Flavour'])
y = df.drop(columns = ['Age','Gender'])
```

```python
x.head(10)
```

```python
y.head(10)
```

```python
from sklearn.tree import DecisionTreeClassifier
Model = DecisionTreeClassifier()
Model.fit(x.values,y.values)
```

```python
age = 17
gender = Encd.transform(['Male'])
Model.predict([[age,gender]])
```

```python
def predict():
  age = input('Enter Age: ')
  g = input('Enter Gender: ')
  gender = Encd.transform([g])
  flavor = Model.predict([[age,gender]])
  print("Recommended Flavor : ", flavor[0])
```
